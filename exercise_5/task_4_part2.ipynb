## Task 4 Part 2: Time delay embedding: Chaotic dynamics


```python
# Imports 

import numpy as np 
from matplotlib import pyplot as plt 
from sw.lorenz import * 

%load_ext autoreload
%autoreload 2 
```

### Lorenz system 

```python
SIM0 = [10.0, 10.0, 10.0]
t = np.arange(0,100,0.01)

#Solving the system with odeint
states = solver(SIM0, t)
```

```python
#Plotting
fig = plt.figure(figsize=(16, 8))
ax0 = fig.add_subplot(1, 2, 1, projection='3d')
ax0.plot(states[:, 0], states[:, 1], states[:, 2], color="g", alpha=0.7)
ax0.set_xlabel('X')
ax0.set_ylabel('Y')
ax0.set_zlabel('Z')
plt.savefig("figures/task_4_2/lorenz.png")
```

### Time delay approximation with X-coordinate 

```python
for i in range(2,21,2):
    dt = 1 * i
    x_0, x_1, x_2 = time_delay_lorenz(states[:,0], dt)
    fig = plt.figure(figsize=(16, 8))
    ax = fig.add_subplot(1, 2, 1, projection='3d')
    ax.plot(x_0, x_1, x_2)
    ax.set_title(f"delta t = {dt}")
    ax.set_xlabel('X0')
    ax.set_ylabel('X1')
    ax.set_zlabel('X2')
    plt.savefig(f"figures/task_4_2/x_embed_dt={dt}")
```

## Time delay approximation with Z-coordinate

```python
for i in range(2,21,2):
    dt = 1 * i
    z_0, z_1, z_2 = time_delay_lorenz(states[:,2], dt)
    fig = plt.figure(figsize=(16, 8))
    ax = fig.add_subplot(1, 2, 1, projection='3d')
    ax.set_title(f"delta t={dt}")
    ax.plot(z_1, z_2,z_0)
    ax.set_xlabel('Z_0')
    ax.set_ylabel('Z_1')
    ax.set_zlabel('Z_2')
    plt.savefig(f"figures/task_4_2/z_embed_dt={dt}")
```
