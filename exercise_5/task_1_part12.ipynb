## Part 2: Linear approximation of nonlinear data set

```python
# imports
from sw.function_approx import * 
from sw.plotting import *
```

## Load data set:

```python
x_lin, F_lin = load_txt_data("datasets/linear_function_data.txt")
x_nlin, F_nlin = load_txt_data("datasets/nonlinear_function_data.txt")
```

## Least-square minimalization:
<img src="https://latex.codecogs.com/svg.image?\hat{A}^{T}=\left(X^{T}&space;X\right)^{-1}&space;X^{T}&space;F" title="\hat{A}^{T}=\left(X^{T} X\right)^{-1} X^{T} F" />

We want to fit the best linear approximation:
<img src="https://latex.codecogs.com/svg.image?f_{\text&space;{linear&space;}}(x)=A&space;x&space;\in&space;\mathbb{R}^{d}" title="f_{\text {linear }}(x)=A x \in \mathbb{R}^{d}" />
with zero offset



```python
A_lin = least_square(x_lin,F_lin,len(x_lin))
A_nlin = least_square(x_nlin,F_nlin,len(x_nlin))
```

# Visualization using matplotlib:

```python
plot_approximation(x_lin,F_lin,A_lin*x_lin,"Linear approximation of linear data set")
plot_approximation(x_nlin,F_nlin,A_nlin*x_nlin,"Linear approximation of nonlinear data set")
```
