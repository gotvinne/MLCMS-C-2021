## Part 3:  
## Using a combination of radial basis function to approximate a non-linear data set

```python
# imports
from sw.function_approx import *
from sw.plotting import *
import matplotlib.pyplot as plt
```

## Load data set:

```python
nonlinear_filepath = "datasets/nonlinear_function_data.txt"
x, F = load_txt_data(nonlinear_filepath)
```

## Radial basis approximation: 

```python
epsilon = 0.3
L = 20

phi = rbf_matrix(x,L,epsilon)
approx, radial_func = rbf_approx(x,F,phi,epsilon)
```

## Visualisation of radial basis function and approximation:

```python
plot_rbf(x,phi,"Radial basis functions")
plot_rbf(x,radial_func,"RBF Composition")
plot_approximation(x,F,approx,"Radial basis approximation","b*")
```
