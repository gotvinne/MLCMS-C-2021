## Plotting phase diagram with streamplot

```python
# These are representative bifurcation values for all bifurcations to the linear system
hyperbolic_bifurcation_values = [-1.8,-1,8.3,0.0,-3]
```

```python
for alpha in hyperbolic_bifurcation_values:
    A = np.array([[3*alpha+1,3],[-3,0.5*alpha*alpha]])
    e_val, _ = eig(A)
    t1.plot_phase_portrait(A,e_val,alpha)
    
```
