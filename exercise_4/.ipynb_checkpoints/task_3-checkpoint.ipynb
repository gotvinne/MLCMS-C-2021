### Training a Variational Autoencoder on MNIST


```python
    #imports 
import numpy as np

import tensorflow as tf
from tensorflow import keras as tfk
import matplotlib.pyplot as plt

#import tensorflow_probability as tfp

# assign probablity distributions variable
#tfd = tfp.distributions
tfkl = tf.keras.layers
#tfpl = tfp.layers
print(tfk.__version__)
print(tf.__version__)
```

```python
# Download MNIST dataset
#mnist = tf.keras.datasets.mnist.load_data()
(trainX, trainY), (testX, testY) = tf.keras.datasets.mnist.load_data()
print('Train', trainX.shape, trainY.shape)
print('Test', testX.shape, testY.shape)
# summarize pixel values
print('Train', trainX.min(), trainX.max(), trainX.mean(), trainX.std())
print('Test', testX.min(), testX.max(), testX.mean(), testX.std())
datagen = tfk.preprocessing.image.ImageDataGenerator(rescale=1.0/255.0)
```

```python
# reshape dataset to have a single channel
x_train = trainX.astype('float32')
x_test = testX.astype('float32')
x_train = x_train / 255
x_test = x_test / 255

# Reshape 
img_width  = x_train.shape[1]
img_height = x_train.shape[2]
num_channels = 1 #MNIST --> grey scale so 1 channel
x_train = x_train.reshape(x_train.shape[0], img_height, img_width, num_channels)
x_test = x_test.reshape(x_test.shape[0], img_height, img_width, num_channels)
'''
width, height, channels = trainX.shape[1], trainX.shape[2], 1
trainX = trainX.reshape((trainX.shape[0], width, height, channels))
testX = testX.reshape((testX.shape[0], width, height, channels))
'''
```

```python
latent_dimention = 2
input_shape= (img_height,img_width,num_channels)
```

```python
def sample_z(args):
  z_mu, z_sigma = args
  eps = tfk.backend.random_normal(shape=(tfk.backend.shape(z_mu)[0], tfk.backend.int_shape(z_mu)[1]))
  return z_mu + tfk.backend.exp(z_sigma / 2) * eps
```

```python
input_img = tfkl.Input(shape=input_shape, name="encoder_input")
x = tfkl.Dense(256, activation=tf.nn.relu)(input_img)
x = tfkl.Dense(256, activation=tf.nn.relu)(x)
conv_shape = tfk.backend.int_shape(x) #Shape of conv to be provided to decoder
x = tfkl.Flatten()(x)


z_mu = tfkl.Dense(latent_dimention, name='latent_mu')(x)
z_sigma = tfkl.Dense(latent_dimention, name='latent_sigma')(x)
z = tfkl.Lambda(sample_z, output_shape=(latent_dimention, ), name='z')([z_mu, z_sigma])
```

```python
encoder = tfk.Model(input_img, [z_mu,z_sigma,z], name="encoder")
print(encoder.summary())

'''
encoder = tfk.Sequential(
[
    tfkl.InputLayer(input_shape=(28, 28, 1)),
    x = tfkl.Dense(256, activation=tf.nn.relu),
    print(x)
    x = tfkl.Dense(256, activation=tf.nn.relu),
    print(x)
    x = tfkl.Flatten(),
    print(x)
]
)
'''
```

```python
decoder_input = tfkl.Input(shape=(latent_dimention, ), name="decoder_input")
```

```python
x = tfkl.Dense(conv_shape[1]*conv_shape[2]*conv_shape[3], activation=tf.nn.relu)(decoder_input)
x = tfkl.Reshape((conv_shape[1],conv_shape[2],conv_shape[3]))(x)
x = tfkl.Dense(256, activation=tf.nn.relu)(x)
x = tfkl.Dense(1, activation=tf.nn.relu)(x)


decoder = tfk.Model(decoder_input, x, name='decoder')
decoder.summary()
```

```python
z_decoded = decoder(z)

class CustomLayer(tfkl.Layer):

    def vae_loss(self, x, z_decoded):
        x = tfk.backend.flatten(x)
        z_decoded = tfk.backend.flatten(z_decoded)
        
        # Reconstruction loss (as we used sigmoid activation we can use binarycrossentropy)
        recon_loss = tfk.metrics.binary_crossentropy(x, z_decoded)
        
        # KL divergence
        kl_loss = -5e-4 * tfk.backend.mean(1 + z_sigma - tfk.backend.square(z_mu) - tfk.backend.exp(z_sigma), axis=-1)
        return tfk.backend.mean(recon_loss + kl_loss)

    # add custom loss to the class
    def call(self, inputs):
        x = inputs[0]
        z_decoded = inputs[1]
        loss = self.vae_loss(x, z_decoded)
        self.add_loss(loss, inputs=inputs)
        return x

# apply the custom loss to the input images and the decoded latent distribution sample
#y = CustomLayer()([input_img, z_decoded])

vae = tfk.Model(input_img, z_decoded, name='vae')

input_img = tfk.backend.flatten(input_img)
z_decoded = tfk.backend.flatten(z_decoded)
        
# Reconstruction loss (as we used sigmoid activation we can use binarycrossentropy)
recon_loss = tf.keras.metrics.binary_crossentropy(input_img, z_decoded)
kl_loss = -5e-4 * tfk.backend.mean(1 + z_sigma - tfk.backend.square(z_mu) - tfk.backend.exp(z_sigma), axis=-1)
total_loss= recon_loss + kl_loss
vae.add_loss(total_loss)
vae.add_metric(total_loss, name='loss', aggregation='mean')

vae.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),loss=None)
vae.summary()
```

```python
print(tf.__version__)
vae.fit(x_train, None, epochs = 2, batch_size = 32, validation_split = 0.2)
```

```python
#Let us automate this process by generating multiple images and plotting
#Use decoder to generate images by tweaking latent variables from the latent space
#Create a grid of defined size with zeros. 
#Take sample from some defined linear space. In this example range [-4, 4]
#Feed it to the decoder and update zeros in the figure with output.


n = 20  # generate 15x15 digits
figure = np.zeros((img_width * n, img_height * n, num_channels))

#Create a Grid of latent variables, to be provided as inputs to decoder.predict
#Creating vectors within range -5 to 5 as that seems to be the range in latent space
grid_x = np.linspace(-5, 5, n)
grid_y = np.linspace(-5, 5, n)[::-1]

# decoder for each square in the grid
for i, yi in enumerate(grid_y):
    for j, xi in enumerate(grid_x):
        z_sample = np.array([[xi, yi]])
        x_decoded = decoder.predict(z_sample)
        digit = x_decoded[0].reshape(img_width, img_height, num_channels)
        figure[i * img_width: (i + 1) * img_width,
               j * img_height: (j + 1) * img_height] = digit

plt.figure(figsize=(10, 10))
#Reshape for visualization
fig_shape = np.shape(figure)
figure = figure.reshape((fig_shape[0], fig_shape[1]))

plt.imshow(figure, cmap='gnuplot2')
plt.show()  
```
