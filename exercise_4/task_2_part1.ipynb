```python
#imports 
import sw.periodic_data_set as data
from sw.diffusion_map import *

from matplotlib import pyplot as plt
from math import pi
%load_ext autoreload
%autoreload 2
```

### Generating csv file for N = 1000

```python
N = 1000
```

```python
data.generate_csv(N)
```

```python
data_points = data.read_csv()
```

### Perform Diffusion Map algorithm:

```python
eta = 0.05
l = 5
ll, eig_func = diffusion_map(eta,l,data_points)
```

### Plot Eigenfunctions agains tk

```python
t_k = [(2*pi*k)/(N+1) for k in range(N)]
for i in range(l):
    fig = plt.figure()
    plt.plot(eig_func[:,i],t_k)
    plt.xlim([-0.5,0.5])
    plt.title("λl: "+str(np.round(ll[i],3)))
    plt.ylabel("t_k")
    plt.xlabel(f"Eig_func_{i+1}")
```
