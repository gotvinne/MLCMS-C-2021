### Part 3 

```python
#txt file to data frame 

ped_df = pd.read_csv("datasets/data_DMAP_PCA_vadere.txt", delimiter=' ')
ped_data = ped_df.to_numpy()
print(ped_data.shape)
```

```python
#Plotting the pedestrians' path 

pedestrians = [1,2]
draw_ped_path(ped_data,pedestrians)

```

```python
### PERFOMING PCA 
u,s,vt,sigma = pca(ped_data,2)
reconstruction = u@sigma@vt
print("energy with 2 components:" ,np.sum(energy_pca(ped_data,2))) 

draw_ped_path(reconstruction,pedestrians, 2)
```

```python
### PERFOMING PCA WITH 3 COMPONENTS
u,s,vt,sigma = pca(ped_data,3)
reconstruction = u@sigma@vt
print("energy with 3 components:" ,np.sum(energy_pca(ped_data,3))) 

draw_ped_path(reconstruction, pedestrians,3)
```
