# Task 2: Diffusion Maps

## Part 3: Trajectories
In this part we will examine the trajectories given in data_DMAP_PCA_vadere.txt

```python
# imports
from sw.diffusion_map import diffusion_map, diffusion_space_mapping
from sw.pca import draw_ped_path
import pandas as pd

from matplotlib import pyplot as plt
```

## Read trajectory data

```python
ped_df = pd.read_csv("datasets/data_DMAP_PCA_vadere.txt", delimiter=' ',header=None)
trajectories = ped_df.to_numpy()
```

## Plot the paths

```python
pedestrians = [1,2]
draw_ped_path(trajectories,pedestrians)
```

## Diffusion map

```python
l = 10
eta = 0.05
ll,eig_func = diffusion_map(eta,l,trajectories)
print(eig_func.shape)
```

```python
for i in range(1,l):
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.scatter(eig_func[:,0], eig_func[:,i])
    plt.xlabel("Phi_1")
    plt.ylabel(f"Phi_{i+1}")
    plt.title("Visualizing - eigenfunctions of trajectory data set")
```
